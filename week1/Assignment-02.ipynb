## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv("company_sales_data.csv") 


# Set the x axis label of the current axis.
plt.xlabel('Month Number')
# Set the y axis label of the current axis.
plt.ylabel('Total Profit')

#Show all profit
#Line style , color and marker
plt.plot(data.total_profit, color='red', linestyle='--', marker='o',markerfacecolor='k',linewidth=3)

#Show Legened
plt.legend(["Total Profit"], loc = 4)

```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("company_sales_data.csv") 


# Set the x axis label of the current axis.
plt.xlabel('Month Number')
# Set the y axis label of the current axis.
plt.ylabel('Total Profit')

#Show all profit
#Line style , color and marker

plt.plot(data.facecream,  marker='o', linewidth=3 ,label='Face cream sales')
plt.plot(data.facewash,     marker='o', linewidth=3, label='Facewash sales')
plt.plot(data.toothpaste,marker='o', linewidth=3, label='Toothpaste sales')
plt.plot(data.bathingsoap, marker='o', linewidth=3 ,label='Bathingsoap sales')
plt.plot(data.shampoo, marker='o', linewidth=3 ,label='Shampoo sales')
plt.plot(data.moisturizer, marker='o', linewidth=3, label='Moisturizer sales')
plt.legend(loc='upper center')


```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("company_sales_data.csv") 

x = data.month_number.tolist()
y = data.toothpaste.tolist()

# Set the x axis label of the current axis.
plt.xlabel('Month Number')
# Set the y axis label of the current axis.
plt.ylabel('Number of Sales')

plt.scatter(x,y, label='Toothpaste sales')
plt.legend(loc='upper left')
plt.grid(True, linewidth= 1, linestyle=":")
plt.show()
```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("company_sales_data.csv") 
months = data.month_number.tolist()
facewash = data.facewash.tolist()
facecream = data.facecream.tolist()

plt.xlabel('Month Number')
plt.ylabel('Number of Sales')
plt.bar(months, facewash, width= 0.25, label = 'Facewash sales', align='edge', color="red")
plt.bar(months, facecream, width= -0.25, label = 'Facecream sales', align='edge', color="green")
plt.legend(loc='upper left')
plt.show()



```

## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 

```python
data['toothpaste'].sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("company_sales_data.csv") 
labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
months = data.month_number.tolist()
sales = [data.facecream.sum(),data.facewash.sum(),data.toothpaste.sum(),data.bathingsoap.sum(),data.shampoo.sum(),
        data.moisturizer.sum()]
plt.pie(sales, labels=labels, autopct='%1.1f%%',)
plt.legend(loc='upper left')
plt.axis("equal")
```
